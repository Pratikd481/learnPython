# Replace and Interpolate

```python
import pandas as pd

csv = pd.read_csv('new.csv')
csv.head()
```

```python
csv.replace(to_replace=0, value=22, inplace=True)
csv
```

```python
csv.replace("[A-Za-z]", "python",regex=True) #replace all alphabets with python
csv
```

```python
csv.replace({"STATUS":"[A-Z]"}, 22,regex=True) #replace with dictionary
csv.head()
```

```python
csv.replace(22,method='bfill') #backward fill
csv.head()
```
