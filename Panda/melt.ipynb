# Melt

```python
# Melt example

import pandas as pd

df = pd.DataFrame({'name':['John', 'Smith', 'Paul', 'Mark'],'english':[80, 70, 90, 85],'math':[90, 85, 88, 90]})
df
```

```python
pd.melt(df)
```

```python
pd.melt(df, id_vars=['name'])
```

```python
pd.melt(df, id_vars=['name'], value_vars=['english']) # This will not work
```

The code pd.melt(df, id_vars=['name'], value_vars=[80]) will not work because of the following reasons:

1. value_vars must contain column names or indices
The value_vars parameter in pd.melt() specifies which columns from the DataFrame to unpivot (transform from wide to long format).
In the given code, [80] is provided as the value for value_vars. If 80 is neither a column name nor a valid index in the DataFrame, pandas will raise a KeyError because it cannot find a column corresponding to 80.
2. value_vars expects iterable column identifiers
Typically, value_vars should be a list of valid column names or column indices (if df.columns are integers).
Using [80] implies that 80 is either:
A column name in the DataFrame (for string-based column labels).
An integer column index (for integer-based column labels).
If 80 does not exist as a valid column in the DataFrame, the operation fails.
